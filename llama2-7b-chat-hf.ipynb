```python
from transformers import AutoTokenizer
import transformers
import torch

model = "meta-llama/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    # device=0,
    # bat
    device_map="auto",
)
```

```python
from datasets import load_dataset

dataset = load_dataset("jfleg", split= 'validation')
```

```python
dataset[0]
```

### create prompt structure

```python
system_prompt = "You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe.  Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content. Please ensure that your responses are socially unbiased and positive in nature. \n\n If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information."
```

```python
base_prompt= "Carefully review the following sentence below and correct any grammar, punctuation, and spelling mistakes you identify. Reply with the corrected sentence: "
```

```python
prompt = '<s>[INST] <<SYS>>' + '\n' + system_prompt + '\n' + '<</SYS>>' + '\n\n' + base_prompt + 'So I think we can not live if old people could not find siences and tecnologies and they did not developped . ' +' [/INST]\n'
```

```python
print(prompt)
```

```python
from tqdm.auto import tqdm

n = 100
for i in tqdm(range(n)):

    # print incorrect text
    print(f"Incorrect text: {dataset[i].get('sentence')}")

    # create prompt body
    prompt = '<s>[INST] <<SYS>>' + '\n' + system_prompt + '\n' + '<</SYS>>' + '\n\n' + base_prompt + dataset[i].get('sentence') +' [/INST]\n'

    sequences = pipeline(
        prompt,
        do_sample=True,
        temperature=1.0,
        top_k=50,
        top_p=0.95,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_new_tokens=1024,
    )
    
    for seq in sequences:
        # print result
        print(f"Result: {seq['generated_text']}")

    print("Correct text:")
    print(dataset[i].get('corrections'))
```

### Create CSV 

```python
from tqdm.auto import tqdm
import pandas as pd

# create an empty dataframe with collumns: sentence, input, output, corrections
results = pd.DataFrame(columns=['sentence', 'input', 'output', 'corrections'])

for i in tqdm(range(100)):
    
    # create prompt body
    prompt = '<s>[INST] <<SYS>>' + '\n' + system_prompt + '\n' + '<</SYS>>' + '\n\n' + base_prompt + dataset[i].get('sentence') +' [/INST]\n'

    sequences = pipeline(
        prompt,
        do_sample=True,
        temperature=1.0,
        top_k=50,
        top_p=0.95,
        num_return_sequences=1,
        eos_token_id=tokenizer.eos_token_id,
        max_new_tokens=1024,
    )

    result_seq = ""
    for seq in sequences:
        result_seq = seq['generated_text']

    # add to dataframe
    results = results.concat({'sentence': dataset[i].get('sentence'), 'input': prompt, 'output': result_seq, 'corrections': dataset[i].get('corrections')}, ignore_index=True)
    
# save dataframe to csv
results.to_csv('dataset.csv', index=False)
```
